```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cpu-full-a.csv", parse_dates=[0], infer_datetime_format=True)  
df.head()
```

```python
df['label'] = False
```

```python
plt.figure(figsize=(20,8))
plt.plot(df['datetime'], df['cpu'], color='teal')
plt.ylabel('CPU %')
plt.title('Time Against CPU Utilization')
plt.show()
```

```python
df.loc[df['datetime'] > pd.Timestamp("2017-01-28 02:00:00"), "label"] = True
```

```python
df.loc[df['datetime'] < pd.Timestamp("2017-01-27 19:00:00"), "label"] = True
```

```python
df.head()
```

```python
plt.figure(figsize=(20,8))

colors = ['red' if x else 'gray' for x in df['label']]
plt.scatter(df['datetime'], df['cpu'], color=colors)
plt.ylabel('CPU %')
plt.title('Time Against CPU Utilization')
plt.show()
```
